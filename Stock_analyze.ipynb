```python
import pandas_datareader.data as web
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df= web.DataReader("TSM","yahoo",start="2016-06-30",end="2021-06-30")
```

```python
#yahoo finance
https://tw.stock.yahoo.com/adr
```

```python
df.head(10)
```

```python
P=df["Adj Close"]
P
```

```python
#價格走勢
P.plot()
```

```python
#diff() 跟前一期的差
P.diff()
```

```python
#報酬率
r= P.diff()/P
r
r.plot()
```

```python
#前100天報酬率變化
r[-100:].plot()
```

```python
#移動平均
MM=P.rolling(window=30).mean()
MMM=P.rolling(window=60).mean()
MMMM=P.rolling(window=100).mean()
```

```python
P[-500:].plot(label="price")
MM[-500:].plot(label="30 days avg")
MMM[-500:].plot(label="60 days avg")
MMMM[-500:].plot(label="100 days avg")
plt.xlabel("Date",fontsize=20,fontweight="bold")
plt.ylabel("Price",fontsize=20,fontweight="bold")
plt.title('Price for TSM in 500 days') #顯示圖表名
plt.legend() #顯示圖標
plt.show()
```
